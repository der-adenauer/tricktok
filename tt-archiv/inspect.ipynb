```python
import pandas as pd
import sqlite3

# Verbindung zur neuen Datenbank herstellen
conn_new = sqlite3.connect('archiv/filtered_tiktok_media.db')

# Daten aus der 'media_info'-Tabelle in einen Pandas DataFrame laden
df = pd.read_sql_query("SELECT * FROM media_info", conn_new)

# Die gewünschten Spalten auf NaN setzen
columns_to_nullify = ['video_id', 'video_uuid', 'embedded_link']
df[columns_to_nullify] = pd.NA

# Den aktualisierten DataFrame zurück in die Datenbank schreiben
# Dabei die alte Tabelle überschreiben
df.to_sql('media_info', conn_new, if_exists='replace', index=False)

# Verbindung zur Datenbank schließen
conn_new.close()

df.head()

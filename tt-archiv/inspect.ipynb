### Filter kein Pfad

````python
import pandas as pd
import sqlite3

# Verbindung zur originalen Datenbank herstellen
conn = sqlite3.connect('archiv/filtered_tiktok_media.db')

# Daten aus der 'media_info'-Tabelle in einen Pandas DataFrame laden
df = pd.read_sql_query("SELECT * FROM media_info", conn)

# Filterkriterien: Entferne alle Zeilen, bei denen 'screenshot_size' 0 ist
filtered_df = df[df['screenshot_size'] != 0]

# Verbindung zur originalen Datenbank schlieÃŸen
conn.close()

# Verbindung zu einer neuen (oder der gleichen um dieselbe Tabelle zu Ã¼berschreiben) Datenbank herstellen
conn_new = sqlite3.connect('archiv/filtered_tiktok_media_new.db')

# Den gefilterten DataFrame in die Datenbank speichern, Tabelle Ã¼berschreiben wenn nÃ¶tig
filtered_df.to_sql('media_info', conn_new, if_exists='replace', index=False)

# Verbindung zur neuen Datenbank schlieÃŸen
conn_new.close()

# Den gefilterten DataFrame anzeigen, um die Ã„nderungen zu Ã¼berprÃ¼fen
filtered_df
````
